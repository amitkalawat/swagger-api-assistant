```python
import base64
from PIL import Image
import io


# Open the image file
image_file = Image.open("uml_diagrams/petstore.jpg")

# Convert the image to bytes
image_bytes = image_file.tobytes()

# Encode the bytes to base64
base64_encoded = base64.b64encode(image_bytes)

# Decode the base64 bytes to a string
base64_string = base64_encoded.decode("utf-8")

# Print the base64 string
print(base64_string)
```

```python
from PIL import Image
import io
import base64

# Load the image
image_path = "uml_diagrams/petstore.jpg" # Replace with the actual path to your image
image = Image.open(image_path)

# Convert the image to RGB mode (if it has an alpha channel)
if image.mode == "RGBA":
    image = image.convert("RGB")

# Convert the image to base64
buffered = io.BytesIO()
image.save(buffered, format="JPEG")
img_bytes = buffered.getvalue()
base64_bytes = base64.b64encode(img_bytes)
base64_string = base64_bytes.decode("utf-8")

# Convert base64 string back to Pillow image
base64_img_bytes = base64_string.encode("utf-8")
base64_img_bytes_decoded = base64.decodebytes(base64_img_bytes)
image_from_base64 = Image.open(io.BytesIO(base64_img_bytes_decoded))

# Display the image
image_from_base64
```
